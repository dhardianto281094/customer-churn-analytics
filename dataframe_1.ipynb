```python
import pandas as pd

# Membaca semua file
df_customer_info = pd.read_csv('Customer_Info.csv')
df_payment_info = pd.read_csv('Payment_Info.csv')
df_service_options = pd.read_csv('Service_Options.csv')
df_online_services = pd.read_csv('Online_Services.csv')
df_status_analysis = pd.read_csv('Status_Analysis.csv')
```

```python
print(df_customer_info.head())
print(df_payment_info.head())
print(df_service_options.head())
print(df_online_services.head())
print(df_status_analysis.head())
```

```python
print(df_customer_info.columns)
print(df_payment_info.columns)
```

```python
def standardize_columns(df):
    df.columns = df.columns.str.strip().str.replace(' ', '').str.replace('-', '').str.replace('_', '').str.capitalize()
    return df

df_customer_info = standardize_columns(df_customer_info)
df_payment_info = standardize_columns(df_payment_info)
df_service_options = standardize_columns(df_service_options)
df_online_services = standardize_columns(df_online_services)
df_status_analysis = standardize_columns(df_status_analysis)

# Merge pertama: Customer Info + Payment Info
# Change 'CustomerID' to the standardized column name, likely 'Customerid'
df_merged = pd.merge(df_customer_info, df_payment_info, on='Customerid', how='inner')

# Merge kedua: hasil merge + Service Options
# Change 'CustomerID' to the standardized column name, likely 'Customerid'
df_merged = pd.merge(df_merged, df_service_options, on='Customerid', how='inner')

# Merge ketiga: hasil merge + Online Services
# Change 'CustomerID' to the standardized column name, likely 'Customerid'
df_merged = pd.merge(df_merged, df_online_services, on='Customerid', how='inner')

# Merge keempat: hasil merge + Status Analysis
# Change 'CustomerID' to the standardized column name, likely 'Customerid'
df_merged = pd.merge(df_merged, df_status_analysis, on='Customerid', how='inner')
```
